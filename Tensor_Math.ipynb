# Tensor Math Operation

*   Add, Subtract, Multiply, Divide, Remianders, Exponents
*   Shorthand and Longhand
*   Reassignment





```python
import torch
import numpy as np
```

```python
np1 = np.random.rand(2,3)
np1
```

```python
tensor_2d = torch.tensor(np1)
tensor_2d
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])

print(tensor_a+tensor_b)
```

```python
# Addition Longhand
torch.add(tensor_a,tensor_b)
```

```python
# Subtraction
tensor_b - tensor_a
```

```python
# Sub Function
torch.sub(tensor_b, tensor_a)
```

```python
# Multiplication
tensor_a * tensor_b
```

```python
# Mul longhand
torch.mul(tensor_a, tensor_b)
```

```python
# Division
tensor_b/tensor_a
```

```python
# Div Longhand
torch.div(tensor_b, tensor_a)
```

```python
# Remainder Modulus
tensor_b % tensor_a
```

```python
# Remainder Longhand
torch.remainder(tensor_b, tensor_a)
```

```python
# Exponents
tensor_a ** tensor_b
```

```python
# another way to write longhand
# tensor_a.add(tensor_b)
```

```python
# Reassignment
tensor_a + tensor_b
```
