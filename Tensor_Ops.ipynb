```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

### Reshape and View

```python
my_torch = my_torch.reshape(2,5)
my_torch
```

```python
# Reshape If we dont the know the number of Items using -1
my_torch2 = torch.arange(10)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(2,-1)
my_torch2
```

```python
my_torch3 = my_torch2.reshape(-1, 5)
my_torch3
```

```python
my_torch4 = torch.arange(10)
my_torch4
```

```python
my_torch4 = my_torch4.view(2,5)
my_torch4
```

```python
# With Reshape and View, They will update
my_torch5 = torch.arange(10)
my_torch5
```

```python
my_torch6 = my_torch5.reshape(2,5)
my_torch6
```

```python
my_torch5[1] = 80085
my_torch5
```

```python
my_torch6
```
